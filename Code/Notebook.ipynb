Create figure

```python
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import sem

def plot_exp(df, axs):
    # Calculate means and SEM for error bars
    grouped = df.groupby(['condition', 'alignment', 'identity'])['distance']
    means = grouped.mean().reset_index()
    errors = grouped.apply(sem).reset_index(name='sem')

    # Merge means and errors
    means = means.merge(errors, on=['condition', 'alignment', 'identity'])

    conditions = ['upright', 'inverted']
    
    for i, condition in enumerate(conditions):
        ax = axs[i]
        condition_df = means[means['condition'] == condition]
        
        # Plot bar plot with error bars
        sns.barplot(x='alignment', y='distance', hue='identity', data=condition_df, palette='muted', ax=ax)

        # Add error bars
        for j, bar in enumerate(ax.patches):
            error = condition_df.iloc[j]['sem']
            x = bar.get_x() + bar.get_width() / 2
            ax.errorbar(x, bar.get_height(), yerr=error, fmt='k', capsize=5)

        # Set titles and labels
        ax.set_ylabel('Mean Distance')
        ax.set_xlabel('Alignment')
        ax.set_xticklabels(['Aligned', 'Misaligned'])
        ax.legend(title='Identity')

        # set text size
        ax.tick_params(axis='both', which='major', labelsize=14)

        # remove spines
        ax.spines['top'].set_visible(False)
        ax.spines['right'].set_visible(False)
```

```python
# fig1
vgg_df = vgg_exp.df
resnet_df = resnet_exp.df
senet_df = senet_exp.df

fig, axs = plt.subplots(3, 2, figsize=(20, 20))
plot_exp(vgg_df, axs[0])
plot_exp(resnet_df, axs[1])
plot_exp(senet_df, axs[2])

# column titles
fig.text(0.3, 0.92, 'Inverted', ha='center', va='center', fontsize=24)
fig.text(0.72, 0.92, 'Upright', ha='center', va='center', fontsize=24)

# row titles
fig.text(0.075, 0.76, 'VGG', ha='center', va='center', rotation='vertical', fontsize=24)
fig.text(0.075, 0.49, 'ResNet', ha='center', va='center', rotation='vertical', fontsize=24)
fig.text(0.075, 0.23, 'SENet', ha='center', va='center', rotation='vertical', fontsize=24)

plt.savefig('psyseminar_fig1.png')
```

```python
# fig2
clip_df = clip_exp.df
featurenet_df = featurenet_exp.df

fig, axs = plt.subplots(3, 2, figsize=(20, 15))
plot_exp(clip_df, axs[0])
plot_exp(featurenet_df, axs[1])

# column titles
fig.text(0.3, 0.92, 'Inverted', ha='center', va='center', fontsize=20)
fig.text(0.72, 0.92, 'Upright', ha='center', va='center', fontsize=20)

# row titles
fig.text(0.075, 0.76, 'CLIP', ha='center', va='center', rotation='vertical', fontsize=20)
fig.text(0.075, 0.49, 'FeatureNet', ha='center', va='center', rotation='vertical', fontsize=20)

plt.savefig('psyseminar_fig2.png')
```
